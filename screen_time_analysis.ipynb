```python
# Import the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Get the dataset from google drive
data = pd.read_csv('drive/MyDrive/Projects/screen_time_analysis/Screentime-App-Details.csv')
data.head()
```

```python
# Change the column names to lower case and replace ' '(space) with '_'(underscore)
data.columns = [col.replace(' ', '_') for col in data.columns.str.lower()]
```

```python
# Check bottom 5 rows
data.tail()
```

```python
# Check the length of data
len(data)
```

```python
# See if our data has any null value
data.isnull().sum()
```

```python
# Check the info of our data
data.info()
```

```python
data.describe()
```

```python
# View value counts of App
data['app'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(13, 6)})
sns.barplot(x='date', y='usage', data=data, hue='app', palette='Dark2')
plt.title('App Usage')
plt.xlabel('Date')
plt.ylabel('Usage')
plt.xticks(rotation=90)
plt.show()
```

From analysing the above chart, we can say that the whatsapp has more usage than instagram

```python
sns.set(rc={'figure.figsize':(13, 6)})
sns.barplot(x='date', y='notifications', data=data, hue='app', palette='Dark2')
plt.title('App Notifications')
plt.xlabel('Date')
plt.ylabel('Notifications')
plt.xticks(rotation=90)
plt.show()
```

From analysing the above chart, we can say that the whatsapp has more notifications than instagram, so it obviously has more usage.

Now let's see how many times applications are open

```python
sns.set(rc={'figure.figsize':(13, 6)})
sns.barplot(x='date', y='times_opened', data=data, hue='app', palette='Dark2')
plt.title('App Times Opened')
plt.xlabel('Date')
plt.ylabel('Times Opened')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(13, 6)})
# sns.scatterplot(x='notifications', y='usage', data=data, size='notifications')
sns.regplot(x='notifications', y='usage', data=data, ci=None)
plt.title('Notifications vs Usage')
plt.xlabel('Notifications')
plt.ylabel('Usage')
plt.show()
```

**Summary:** The analysis of the above charts reveals that WhatsApp has been used more than Instagram, based on the usage, number of notifications and time opened. 

This shows that a higher number of notifications leads to increased usage, indicating a direct correlation between the two. This indicates that applications with more notifications are likely to have higher usage.
